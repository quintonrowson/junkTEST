```python
data_file='/content/drive/My Drive/python/Mariya_youtube_chan/L4/SMSSpamCollection.txt'
```

```python
import pandas as pd
data=pd.read_csv(data_file,sep='\t',header=None,names=['label','SMS'])
data.head()
```

```python
#load stop words and punctuation
import string
import nltk
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('punkt')

stopwords=nltk.corpus.stopwords.words('english')
punctuation=string.punctuation
print(stopwords[:5])
print(punctuation)
```

```python
#pre-process sms content
def preprocess(SMS):
  lowercase=SMS.lower()
  tokenize=nltk.tokenize.word_tokenize(lowercase)
  remove_punctuation=[word for word in tokenize if word not in punctuation]
  #qt=nltk.tokenize.word_tokenize(remove_punctuation)
  remove_stopwords=[word for word in remove_punctuation if word not in stopwords]

  return remove_stopwords

data['processed']=data['SMS'].apply(lambda x:preprocess(x))
data.head()

```

```python
#catagorising spam/ham associated messages
def categorize_words():
  spam_words=[]
  ham_words=[]
  for SMS in data['processed'][data['label']=='spam']:
    for word in SMS:
      spam_words.append(word)


  for SMS in data['processed'][data['label']=='ham']:
    for word in SMS:
      ham_words.append(word)

  return spam_words,ham_words

spam_words,ham_words = categorize_words()

#check how many words and what type of words are in span an ham
print("There are",spam_words.count('free'),"spam words and include:",spam_words[:5])
print("")
print("There are",ham_words.count('free'),"spam words and include:",ham_words[:5])
```

```python
# iterate over all the words from the user input and count their occurance in both the ham words and spam words
def predict(user_input):
  spam_counter=0
  ham_counter=0

  for word in user_input:
    spam_counter+=spam_words.count(word)
    ham_counter+=ham_words.count(word)

  print("****************RESULTS************************************")
  if ham_counter>spam_counter:
    #add accuracy
    accuracy = round((ham_counter/(spam_counter+ham_counter))*100,2)
    print("The message is NOT spam with {}% accuracy".format(accuracy))
  elif spam_counter>ham_counter:
    #add accuracy
    accuracy = round((spam_counter/(spam_counter+ham_counter))*100,2)
    print("The message is SPAM with {}% accuracy".format(accuracy))
  else:
    print("This message cound be spam with 50% accuracy")

  print("**********************************************************")

```

```python
# collect user input
user_input = input('Please enter a spam or ham messae to see if our function predicts properly\n')
```

```python
process_input=preprocess(user_input)
predict(process_input)
```
